```python
# Util print table
def print_table(table):
    title=['pos_inicial','pos_final','match']
    len_match = table[0][1] - table[0][0] + 2 if table else 50
    print('{1:^13}│{2:^11}│{3:^{0}}'.format(len_match, *title))
    print('─'*13+'┼'+'─'*11+'┼'+'─'*(len_match))
    for pos_inicial, pos_final in table:
        print(f'{pos_inicial:>12} │{pos_final:>10} │{sequence[pos_inicial:pos_final+1]:>{len_match}}')
    
```

Ejemplo Bioinformático de busqueda por metodo de fuerza bruta


```python
def fuerza_bruta(target, sequence):
    target_len = len(target)
    sequence_len = len(sequence)

    result = []
    # se itera sobre la secuecia (teniendo cuidado de no desbordarse)
    for i in range(sequence_len - target_len + 1):
        # se itera sobre la cadena objetivo
        for j, _ in enumerate(target): # j es el indice _ es la letra actual en target (target[j])
            if (target[j] == sequence[i + j]): # si la letra coincide con la secuencia sigue iterando la cadena objetivo
                if (j < target_len - 1): # si no es la ultima iteracion entonces continue
                    continue
                result.append((i, i + target_len - 1)) # adjunta el indice inicial y final donde se encuentra el objetivo al result
                break
            break # si no coincide deja de iterar sobre la cadena objetivo

    return result # en caso tal que no se encuentre el objetivo en la secuencia retorna []

target='TTGAAAGAGCCACCACATTTT'
# target='ACAGTAAGTGACAACAGATGTTT' # Ejemplo profesor
print_table(fuerza_bruta(target, sequence))
```
